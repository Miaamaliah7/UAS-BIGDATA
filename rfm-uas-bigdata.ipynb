# a. Bentuklah RFM (Recency Frequency Monetery) model untuk segmentasi pengunjung Giant 

```python
df= pd.read_csv('../input/rfm-uas-data/UAS_RFM.csv')
df.head (10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['TrxDate'] = pd.to_datetime(df['TrxDate'])
```

```python
max_date = max(df['TrxDate'])
max_date
```

```python
df['Diff'] = max_date - df['TrxDate']
df.head()
```

```python
type_counts = df['CardID'].value_counts()
CardID=pd.DataFrame(type_counts)
CardID.head()
```

```python
RFM = df.groupby('CardID')['Diff'].min().reset_index()
RFM.head()
```

```python
RFM['Diff'] = RFM['Diff'].dt.days
RFM.columns = ['CardID','Recency']
RFM.head()
```

```python
RFM_F = df.groupby('CardID')['TrxDate'].count().reset_index()
RFM_F.columns = ['CardID', 'Frequency']
RFM_F.head()

```

```python
RFM_F.tail(5)
```

```python
RFM_M = df.groupby('CardID')['Amount'].sum().reset_index()
RFM_M.columns = ['CardID','Amount']
RFM_M.head()
```

```python
RFM_M.tail()
```

```python
RFM = RFM.merge(RFM_F,how='inner',on=['CardID'])
RFM =RFM.merge(RFM_M,how='inner',on=['CardID'])
RFM.head(25)
```

```python
RFM.tail(25)
```
